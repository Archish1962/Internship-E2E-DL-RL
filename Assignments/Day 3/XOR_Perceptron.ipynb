Implement a simple single/Multi-layer perceptron from scratch in Python to solve a basic binary classification problem (e.g., XOR gate). Train, it using gradient descent and visualize its decisionÂ boundary.

Here a multi-layer perceptron is implemented for solving XOR Gate classifiction problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# Plotting decision boundary
def plot_decision_boundary(W1, b1, W2, b2, X, y, title="Decision Boundary"):
    xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 200), np.linspace(-0.5, 1.5, 200))
    grid = np.c_[xx.ravel(), yy.ravel()]
    
    z1 = np.dot(grid, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    y_pred = sigmoid(z2)
    y_pred = y_pred.reshape(xx.shape)

    plt.figure(figsize=(6, 5))
    plt.contourf(xx, yy, y_pred, levels=[0, 0.5, 1], colors=["#FFAAAA", "#AAAAFF"], alpha=0.6)
    
    # Class 0
    plt.scatter(X[y.ravel()==0][:, 0], X[y.ravel()==0][:, 1], c='red', edgecolor='k', label="Class 0")
    # Class 1
    plt.scatter(X[y.ravel()==1][:, 0], X[y.ravel()==1][:, 1], c='blue', edgecolor='k', label="Class 1")

    plt.title(title)
    plt.xlabel("Input 1")
    plt.ylabel("Input 2")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# XOR truth table
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([[0], [1], [1], [0]])

```

```python
# These are the hyperparameters
np.random.seed(42)

input_dim = 2
hidden_dim = 4
output_dim = 1
learning_rate = 0.2
epochs = 5000

# Weights and biases
W1 = np.random.randn(input_dim, hidden_dim)
b1 = np.zeros((1, hidden_dim))
W2 = np.random.randn(hidden_dim, output_dim)
b2 = np.zeros((1, output_dim))
```

```python
losses = []

for epoch in range(epochs):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    y_pred = sigmoid(z2)
    y_class = (y_pred > 0.5).astype(int)



    # Loss (binary cross-entropy)
    loss = np.mean((y - y_pred)**2)
    losses.append(loss)

    # backpropagation
    d_loss = 2 * (y_pred - y)
    d_output = d_loss * sigmoid_derivative(y_pred)

    d_hidden = d_output.dot(W2.T) * sigmoid_derivative(a1)

    # Weights updation
    W1 -= X.T.dot(d_hidden) * learning_rate
    # print(f"W1: {W1}")
    b1 -= np.sum(d_hidden, axis=0, keepdims=True) * learning_rate
    W2 -= a1.T.dot(d_output) * learning_rate
    # print(f"W2: {W2}")
    b2 -= np.sum(d_output, axis=0, keepdims=True) * learning_rate

    # Print loss and output predicted
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
        print("W1:\n", W1)
        print("b1:\n", b1)
        print("W2:\n", W2)
        print("b2:\n", b2)
        plot_decision_boundary(W1, b1, W2, b2, X, y, title=f"Epoch {epoch}")
        print(f"Predicted outputs at epoch {epoch}:\n", np.round(y_pred.ravel(), 3))
        print("\n")


plot_decision_boundary(W1, b1, W2, b2, X, y, title="Final Decision Boundary")
print(f"Predicted outputs at epoch {epoch}:\n", np.round(y_pred.ravel(), 3))

```

```python
print("XOR Truth Table with Predictions:\n")
print("Input 1 | Input 2 | Expected | Predicted | Class")
print("-----------------------------------------------")
for i in range(len(X)):
    x1, x2 = X[i]
    expected = y[i][0]
    pred_prob = y_pred[i][0]
    pred_class = y_class[i][0]
    print(f"   {x1}     |    {x2}    |    {expected}     |   {pred_prob:.3f}   |   {pred_class}")
```
