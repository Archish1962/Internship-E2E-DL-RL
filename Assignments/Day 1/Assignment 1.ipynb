```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the dataset

```python
data = pd.read_csv("C:/Users/archi/OneDrive/Documents/Coding/Python/Assignments/Day 1/Mental health Depression disorder Data.csv", low_memory=False)
data
```

```python
print(data.columns)
```

```python
data = data.rename(columns={"Schizophrenia (%)":"Schizophrenia","Bipolar disorder (%)":"Bipolar","Eating disorders (%)":"Eating", "Anxiety disorders (%)":"Anxiety","Drug use disorders (%)":"Drug","Depression (%)":"Depression", "Alcohol use disorders (%)":"Alcohol"})
print( "data columns: \n" , data.columns )
```

```python
print(data.shape)
```

Data Cleaning and Preprocessing

```python
altered_table = data.drop(["index","Code"], axis=1) #Removal of index and code columns from the dataset
altered_table
```

```python
altered_table.drop_duplicates() #Removing repeated rows from the table
```

```python
data_without_missingvalues = altered_table.dropna(subset=["Anxiety"], inplace=False)
```

```python
data_without_missingvalues #Considering Anxiety disorders column to not have missing values
```

```python
print("Size of the data")
data_without_missingvalues.shape
```

```python
print("max values: \n" , data_without_missingvalues.iloc[:,:].max()) # No unwanted values
print("min values: \n" , data_without_missingvalues.iloc[:,:].min()) # No unwanted values
```

Data Visualization

```python
avg_countries_anxiety_by_year = data_without_missingvalues.dropna(subset=["Anxiety"]).groupby("Year")["Anxiety"].mean().reset_index()
print(avg_countries_anxiety_by_year)
```

```python
plt.figure(figsize=(50,10),dpi=100)
plt.plot(avg_countries_anxiety_by_year["Year"], avg_countries_anxiety_by_year["Anxiety"], marker='o', color='teal', linewidth=2)

plt.title("Average Anxiety from 1990 to 2017", fontsize=18, color="blue")
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Anxiety (%)", fontsize=14)
plt.grid(True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
avg_countries_anxiety = data_without_missingvalues.dropna(subset=["Anxiety"]).groupby("Entity")["Anxiety"].mean().reset_index()
print(avg_countries_anxiety)
```

```python
max_avg_anxiety = avg_countries_anxiety["Anxiety"].max()
print("max anxiety: \n" , avg_countries_anxiety[avg_countries_anxiety["Anxiety"] ==  max_avg_anxiety])
```

```python
min_avg_anxiety = avg_countries_anxiety["Anxiety"].min()
print("min anxiety: \n" , avg_countries_anxiety[avg_countries_anxiety["Anxiety"] ==  min_avg_anxiety])
```

```python
# Scatter plot
plt.figure(figsize=(50,10),dpi=100)
plt.scatter(avg_countries_anxiety["Entity"], avg_countries_anxiety["Anxiety"], color="skyblue", label="Avg Anxiety", edgecolors='black')

# max and min anxiety
max_row = avg_countries_anxiety[avg_countries_anxiety["Anxiety"] == avg_countries_anxiety["Anxiety"].max()]
min_row = avg_countries_anxiety[avg_countries_anxiety["Anxiety"] == avg_countries_anxiety["Anxiety"].min()]

plt.scatter(max_row["Entity"], max_row["Anxiety"], color="red", label="Max Anxiety", s=100, edgecolors='black')
plt.scatter(min_row["Entity"], min_row["Anxiety"], color="green", label="Min Anxiety", s=100, edgecolors='black')

# marking the min and maximum anxiety by
plt.annotate(f"Max: {max_row['Entity'].values[0]}", 
             (max_row["Entity"].values[0], max_row["Anxiety"].values[0]), 
             textcoords="offset points", xytext=(0,10), color='red')

plt.annotate(f"Min: {min_row['Entity'].values[0]}", 
             (min_row["Entity"].values[0], min_row["Anxiety"].values[0]), 
             textcoords="offset points", xytext=(0,-15), color='green')

# Labeling and layout
plt.xticks(rotation=90)
plt.xlabel("Country (Entity)")
plt.ylabel("Average Anxiety")
plt.title("Average Anxiety by Country with Max/Min Highlighted")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Histogram 
plt.figure(figsize=(12,8),dpi=100)
buckets = np.arange(2,9,0.5)
plt.hist(avg_countries_anxiety["Anxiety"], bins=buckets, edgecolor='black')
plt.title("Anxiety Histogram",fontsize=20 ,color="blue")
plt.xlabel("Anxiety %",fontsize=20 ,color="blue")
plt.ylabel("Countries Count",fontsize=20 ,color="blue")
plt.yticks(fontsize=15 ,color ="black")
plt.xticks(fontsize=15 ,color ="black")
plt.grid()
plt.show()
```
